### What This Does

- Researcher Agent finds information
- Analyst Agent processes and analyzes
- Writer Agent creates professional reports

Instead of one AI doing everything,
multiple specialized agents collaborate

```python
import warnings
warnings.filterwarnings('ignore')
import os
import anthropic
from google.colab import userdata

# Setup
client = anthropic.Anthropic(
    api_key=userdata.get("ANTHROPIC_API_KEY")
)

# ============================================
# AGENT RUNNER FUNCTION
# ============================================
def run_agent(role, goal, task):
    """Run a single specialized agent"""
    response = client.messages.create(
        model="claude-haiku-4-5",
        max_tokens=1024,
        system=f"""You are a {role}.
        Your goal is: {goal}
        Be concise, professional and use plain text only.""",
        messages=[{
            "role": "user",
            "content": task
        }]
    )
    return response.content[0].text

# ============================================
# MULTI-AGENT PIPELINE
# ============================================
def multi_agent_system(topic):
    """Run 3 agents in sequence to produce report"""

    print("=" * 50)
    print(f"Multi-Agent System Starting...")
    print(f"Topic: {topic}")
    print("=" * 50)

    # Agent 1 — Researcher
    print("\n🔍 Agent 1: Researcher working...")
    research = run_agent(
        role="Research Analyst",
        goal="Find accurate information and key insights",
        task=f"Research this topic thoroughly and list 5 key findings: {topic}"
    )
    print("Research complete!")

    # Agent 2 — Analyst
    print("\n📊 Agent 2: Analyst working...")
    analysis = run_agent(
        role="Data Analyst",
        goal="Analyze research and extract most important insights",
        task=f"Analyze these research findings and identify the 3 most important insights:\n{research}"
    )
    print("Analysis complete!")

    # Agent 3 — Writer
    print("\n✍️ Agent 3: Writer working...")
    report = run_agent(
        role="Professional Report Writer",
        goal="Write clear professional reports",
        task=f"""Write a professional report based on:

Research Findings:
{research}

Key Analysis:
{analysis}

Format as:
EXECUTIVE SUMMARY
KEY FINDINGS
ANALYSIS
CONCLUSION"""
    )
    print("Report complete!")

    return {
        "research": research,
        "analysis": analysis,
        "report": report
    }

# ============================================
# TEST THE MULTI-AGENT SYSTEM
# ============================================
print("Multi-Agent System Ready!")
print("3 specialized agents will collaborate!\n")

# Test 1
results = multi_agent_system(
    "Benefits of AI agents in business and education"
)

print("\n" + "=" * 50)
print("FINAL REPORT FROM MULTI-AGENT SYSTEM:")
print("=" * 50)
print(results["report"])

print("\n" + "=" * 50)
print("Test 2: Different Topic")
print("=" * 50)

results2 = multi_agent_system(
    "Future of mathematics education with AI"
)
print("\nFINAL REPORT:")
print(results2["report"])
```
